```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Initialize variables to store the best clustering results
best_k = 0
best_labels = None
best_silhouette_score = -1

silhouette_scores = []

# Loop through different values of K (number of clusters)
for k in range(2, 9):
    # Perform K-means clustering with random initialization, using n_init=10
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=1234)
    cluster_labels = kmeans.fit_predict(MD_x)

    # Calculate the silhouette score to evaluate clustering quality
    silhouette_avg = silhouette_score(MD_x, cluster_labels)
    silhouette_scores.append(silhouette_avg)

    # Check if the current K gives a better silhouette score
    if silhouette_avg > best_silhouette_score:
        best_k = k
        best_labels = cluster_labels
        best_silhouette_score = silhouette_avg

# Now, best_k contains the optimal number of clusters.
print("Optimal number of clusters:", best_k)

# You can access the cluster labels through best_labels if needed.
# For example, to print the labels:
print("Cluster Labels:")
print(best_labels)
```

```python
plt.plot(range(2, 9), silhouette_scores, marker='o')
plt.xlabel('Number of Segments (Clusters)')
plt.ylabel('Silhouette Score')
plt.title('K-means Clustering Performance')
plt.xticks(range(2, 9))
plt.grid(True)
plt.show()
```

```python
# Initialize lists to store the information criteria
aic_scores = []
bic_scores = []
icl_scores = []

# Loop through different values of K (number of clusters)
for k in range(2, 9):
    # Perform Gaussian Mixture Model clustering with k components
    gmm = GaussianMixture(n_components=k, random_state=1234)
    gmm.fit(MD_x)
    cluster_labels = gmm.predict(MD_x)

    # Calculate the AIC, BIC, and ICL scores for the current clustering
    aic = gmm.aic(MD_x)
    bic = gmm.bic(MD_x)
    icl = gmm.score(MD_x) - 0.5 * (k * np.log(MD_x.shape[0]))

    # Append the scores to the respective lists
    aic_scores.append(aic)
    bic_scores.append(bic)
    icl_scores.append(icl)

# Plot the information criteria scores
plt.plot(range(2, 9), aic_scores, marker='o', label='AIC')
plt.plot(range(2, 9), bic_scores, marker='o', label='BIC')
plt.plot(range(2, 9), icl_scores, marker='o', label='ICL')
plt.xlabel('Number of Segments (Clusters)')
plt.ylabel('Value of Information Criteria')
plt.title('Model Selection for Gaussian Mixture Models')
plt.xticks(range(2, 9))
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
# Calculate the pairwise distance matrix
dist_matrix = pdist(MD_x.T)

# Perform hierarchical clustering using complete linkage
hclust_result = linkage(squareform(dist_matrix), method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(hclust_result, labels=selected_cols.columns, orientation='top', leaf_font_size=10)
plt.xlabel('Features')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```
